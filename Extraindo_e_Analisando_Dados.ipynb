## Respondendo pergutnas de Negócios com Análise de Dados
Precisamos gerar uma amostra de dados com os pacientes com mais de 50 anos e para cada um deles indicar uma nova coluna se o paciente está normal (BMI menor que 30) ou obeso (BMI maior ou igual a 30). Entã devemos gerar um novo arquivo CSV e encaminhar para o tomador de decisão.

#### Retorne Age, Glucose e Outcome para pacientes com Glucose maior que 195. 

```python
%%sql

SELECT Age, Glucose, Outcome FROM diabetes WHERE Glucose > 195
```

```python
df.columns
```

#### Vamos criar uma nova tabela no banco de dados.

```python
%%sql

CREATE TABLE pacientes (Pregnancies INT, 
                        Glucose INT,
                        BloodPressure INT, 
                        SkinThickness INT,
                        Insulin INT,
                        BMI DECIMAL(8, 2),
                        DiabetesPedigreeFunction DECIMAL(8, 2),
                        Age INT,
                        Outcome INT);
```

#### A tabela está vazia

```python
%%sql

SELECT * FROM pacientes
```

#### Vamos copiar o conteúdo de uma tabela para outra, mas somente para pacientes maiores que 50 anos de idade.

```python
%%sql

INSERT INTO pacientes (Pregnancies,
                        Glucose,
                        BloodPressure,
                        SkinThickness,
                        Insulin,
                        BMI,
                        DiabetesPedigreeFunction,
                        Age,
                        Outcome)
SELECT Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age, Outcome
FROM diabetes
WHERE Age > 50;
```

#### Retorna os pacientes

```python
%%sql

SELECT * FROM pacientes
```

#### Vamos alterar a tabela e incluir uma nova coluna.

```python
%%sql

ALTER TABLE pacientes
ADD Perfil VARCHAR(10);
```

#### Coluna criada:

```python
%%sql

SELECT * FROM pacientes;
```

#### Agora vamos atualizar a coluna com o valor "Normal" se o BMI for menor do que 30. 

```python
%%sql

UPDATE pacientes
SET Perfil = 'Normal'
WHERE BMI < 30;
```

#### Conferimos o resultado:

```python
%%sql

SELECT * FROM pacientes
```

#### Agora vamos atualizar a coluna com o valor "Obeso" se o BMI for maior ou igual a 30.

```python
%%sql

UPDATE pacientes
SET Perfil = 'Obeso'
WHERE BMI >= 30;
```

#### Conferimos o resultado:

```python
%%sql

SELECT * FROM pacientes;
```

## Retornando os Dados para o Pandas e Salvando o CSV

```python
#Query
dsa_query = cnn.execute("SELECT * FROM pacientes")
```

```python
dsa_query
```

```python
# List Comprehension para retornar os metadodos da tabela (nomes de colunas)
cols = [coluna[0] for coluna in dsa_query.description]
```

```python
cols
```

```python
#Gera o dataframe
resultado = pd.DataFrame.from_records(data = dsa_query.fetchall(), columns = cols)
```

```python
#Shape
resultado.shape
```

```python
#Visualiza
resultado.head()
```

```python
#Salva em CSV
resultado.to_csv('resultadofn.csv', index = False)
```
